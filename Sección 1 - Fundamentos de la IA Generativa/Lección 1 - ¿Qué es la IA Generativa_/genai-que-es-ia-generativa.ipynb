## Qué es la inteligencia artificial

La **Inteligencia Artificial** (IA) es el campo de la informática dedicado a crear sistemas capaces de realizar tareas que normalmente requieren inteligencia humana. Estas tareas incluyen reconocer imágenes, comprender lenguaje natural, tomar decisiones y resolver problemas.

A diferencia de la programación tradicional, donde el desarrollador escribe reglas explícitas para cada situación, los sistemas de IA pueden **aprender patrones** y adaptarse a nuevos escenarios. Un programa tradicional para detectar spam requeriría que alguien definiera manualmente todas las palabras sospechosas. Un sistema de IA aprende a identificar spam analizando miles de ejemplos.

> La IA no es una tecnología única, sino un **campo amplio** que engloba múltiples enfoques y técnicas para simular capacidades cognitivas.

```mermaid
flowchart LR
    subgraph Tradicional["Programación tradicional"]
        R[Reglas escritas] --> P1[Programa]
        D1[Datos] --> P1
        P1 --> S1[Salida]
    end
    
    subgraph IAApproach["Enfoque IA"]
        D2[Datos] --> M[Modelo]
        E[Ejemplos] --> M
        M --> S2[Salida aprendida]
    end
    
    style Tradicional fill:#e8f4f8
    style IAApproach fill:#f0e8f8
```

En la programación tradicional, el desarrollador codifica las reglas. En el enfoque de IA, el sistema **descubre las reglas** a partir de los datos. Esta diferencia fundamental es lo que permite que los sistemas de IA manejen problemas demasiado complejos para ser programados manualmente.

## Machine Learning: aprendizaje automático

Dentro del campo de la IA, el **Machine Learning** (ML) o aprendizaje automático es el área que permite a los sistemas aprender de los datos sin ser programados explícitamente para cada tarea. En lugar de escribir instrucciones detalladas, se entrena un modelo con ejemplos.

El proceso de Machine Learning sigue un patrón común:

* Se recopilan **datos de entrenamiento** con ejemplos del problema a resolver.
* Un algoritmo analiza estos datos y ajusta los **parámetros del modelo**.
* El modelo resultante puede hacer **predicciones** sobre datos nuevos que nunca ha visto.

```mermaid
flowchart TB
    subgraph Entrenamiento["Fase de entrenamiento"]
        DE[Datos etiquetados] --> ALG[Algoritmo ML]
        ALG --> MOD[Modelo entrenado]
    end
    
    subgraph Inferencia["Fase de inferencia"]
        DN[Dato nuevo] --> MOD2[Modelo]
        MOD2 --> PRED[Predicción]
    end
    
    MOD -.-> MOD2
    
    style Entrenamiento fill:#e8f4f8
    style Inferencia fill:#f0e8f8
```

> El Machine Learning transforma la forma de resolver problemas: en lugar de programar *cómo* resolver algo, proporcionamos *ejemplos* de lo que queremos y el sistema aprende las reglas por sí mismo.

Los algoritmos de ML tradicionales incluyen árboles de decisión, regresión logística y máquinas de vectores de soporte. Estos algoritmos funcionan bien para problemas con características claras y datos estructurados, pero tienen limitaciones cuando los datos son complejos, como imágenes o texto.

## Deep Learning: aprendizaje profundo

El **Deep Learning** (DL) es una especialización del Machine Learning que utiliza **redes neuronales artificiales** con múltiples capas. Estas capas permiten al modelo aprender representaciones cada vez más abstractas de los datos.

La diferencia clave respecto al ML tradicional está en cómo se extraen las características de los datos:

* En ML tradicional, un experto debe definir qué **características** son relevantes para el problema.
* En Deep Learning, el modelo **aprende automáticamente** las características importantes directamente de los datos crudos.

```mermaid
flowchart TB
    subgraph MLTradicional["ML tradicional"]
        D1[Datos] --> FE[Extracción manual<br/>de características]
        FE --> A1[Algoritmo]
        A1 --> R1[Resultado]
    end
    
    subgraph DeepLearning["Deep Learning"]
        D2[Datos crudos] --> NN[Red neuronal<br/>multicapa]
        NN --> R2[Resultado]
    end
    
    style MLTradicional fill:#e8f4f8
    style DeepLearning fill:#f0e8f8
```

Para reconocer gatos en imágenes, el ML tradicional requeriría que alguien definiera manualmente características como bordes, texturas y formas. Una red neuronal profunda recibe los píxeles directamente y aprende por sí misma qué patrones visuales identifican a un gato.

> El Deep Learning elimina la necesidad de ingeniería manual de características, permitiendo abordar problemas donde definir reglas explícitas sería imposible.

Las redes profundas tienen **múltiples capas ocultas**, cada una aprendiendo representaciones más abstractas. Las primeras capas detectan bordes y texturas simples. Las capas intermedias combinan estos elementos en formas más complejas. Las capas finales reconocen objetos completos.

## IA Generativa: de analizar a crear

La **IA Generativa** representa una especialización del Deep Learning con un objetivo diferente. Mientras que los sistemas de ML tradicionales y Deep Learning se enfocan en *analizar*, *clasificar* o *predecir*, la IA Generativa se centra en **crear contenido nuevo**.

```mermaid
flowchart TB
    IA["Inteligencia Artificial<br/><i>Campo general</i>"] --> ML
    ML["Machine Learning<br/><i>Aprendizaje de datos</i>"] --> DL
    DL["Deep Learning<br/><i>Redes neuronales profundas</i>"] --> GenAI
    GenAI["IA Generativa<br/><i>Creación de contenido</i>"]
    
    style IA fill:#2d3748,color:#fff
    style ML fill:#4a5568,color:#fff
    style DL fill:#718096,color:#fff
    style GenAI fill:#9f7aea,color:#fff
```

Esta jerarquía es fundamental para entender el ecosistema:

* **IA** engloba todo sistema que simula capacidades cognitivas.
* **ML** es el subconjunto que aprende de datos.
* **DL** es la especialización que usa redes neuronales profundas.
* **IA Generativa** aplica Deep Learning para producir contenido nuevo.

> La IA Generativa no es un tipo diferente de IA, sino una **aplicación especializada** del Deep Learning orientada a la síntesis de datos.

### De la predicción a la creación

La distinción entre sistemas tradicionales y generativos es conceptual. Un modelo de Deep Learning para clasificación analiza una imagen y responde: *esto es un gato*. Un modelo generativo puede crear una imagen completamente nueva de un gato que nunca ha existido.

| Característica | ML / DL tradicional | IA Generativa |
|----------------|---------------------|---------------|
| **Objetivo** | Clasificar o predecir | Crear contenido nuevo |
| **Salida** | Etiqueta o valor numérico | Texto, imagen, audio, video, código |
| **Pregunta** | *¿Qué es esto?* | *¿Qué podrías crear?* |

```mermaid
flowchart LR
    subgraph Tradicional["ML / DL tradicional"]
        A[Imagen de gato] --> B[Modelo]
        B --> C["Resultado: gato"]
    end
    
    subgraph Generativa["IA Generativa"]
        D["Prompt: un gato<br/>durmiendo al sol"] --> E[Modelo generativo]
        E --> F[Imagen nueva de gato]
    end
    
    style Tradicional fill:#e8f4f8
    style Generativa fill:#f0e8f8
```

Los modelos generativos funcionan aprendiendo la **distribución estadística** de los datos de entrenamiento. Durante la generación, no copian ni reproducen ejemplos existentes, sino que sintetizan contenido nuevo combinando los patrones aprendidos.

Esta capacidad de síntesis permite crear texto coherente, imágenes realistas, audio natural y código funcional, todo a partir de instrucciones en lenguaje natural. El modelo no busca en una base de datos: calcula qué contenido es más probable dado el contexto proporcionado.
