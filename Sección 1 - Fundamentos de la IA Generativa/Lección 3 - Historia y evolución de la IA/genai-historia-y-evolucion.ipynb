## De la neurona artificial a las redes profundas

La historia de la IA Generativa tiene sus raíces en la búsqueda por imitar el funcionamiento del cerebro humano. En 1943, McCulloch y Pitts propusieron el primer modelo matemático de una **neurona artificial**. Este concepto se materializó en 1958 cuando Frank Rosenblatt desarrolló el **perceptrón**, la primera red neuronal capaz de aprender de los datos.

> El perceptrón demostró que los sistemas artificiales podían aprender patrones, pero su limitación a problemas linealmente separables frenó la investigación durante décadas.

Tras un período de estancamiento conocido como **invierno de la IA**, el punto de inflexión llegó con **backpropagation** (algoritmo que permite entrenar redes multicapa) y la disponibilidad de GPUs. En 2012, **AlexNet** ganó ImageNet con una red neuronal profunda, marcando el inicio de la era del **Deep Learning**.

En 2013, **Word2Vec** revolucionó el procesamiento del lenguaje al representar palabras como vectores donde las similares quedaban cerca. Esta técnica de *embeddings* sentó las bases para que los modelos de lenguaje comprendieran el significado de las palabras.

## Primeras técnicas generativas

Con las bases del Deep Learning establecidas, la investigación se dirigió hacia una pregunta diferente: *¿pueden las redes neuronales crear contenido nuevo en lugar de solo clasificar?*

Los **Variational Autoencoders** (VAE, 2013) fueron los primeros modelos generativos exitosos, comprimiendo datos en un *espacio latente* que permite generar nuevas muestras. En 2014, las **Generative Adversarial Networks** (GANs) revolucionaron la generación de imágenes mediante dos redes en competencia: un generador que crea datos falsos y un discriminador que intenta detectarlos.

> Las GANs funcionan como un falsificador y un detective: el generador mejora creando falsificaciones más convincentes, mientras el discriminador mejora detectándolas.

Mientras tanto, las **redes recurrentes** (RNN) y sus variantes **LSTM** dominaban el procesamiento de secuencias de texto entre 2014 y 2016. Sin embargo, su procesamiento secuencial limitaba tanto la velocidad como la capacidad de capturar dependencias de largo alcance.

## La arquitectura Transformer

El salto definitivo hacia la IA Generativa moderna se produjo en 2017 con *"Attention Is All You Need"*. La arquitectura **Transformer** abandonó las redes recurrentes en favor de un mecanismo nuevo: la **atención**.

El mecanismo de **self-attention** permite que cada elemento de una secuencia *atienda* directamente a todos los demás, calculando su relevancia relativa. Esto significa procesamiento paralelo (aprovechando GPUs) y captura natural de dependencias de largo alcance.

> La atención permite al modelo decidir dinámicamente qué partes de la entrada son relevantes para cada posición de la salida, sin las limitaciones de las redes recurrentes.

### La bifurcación arquitectónica

En 2018, el Transformer se bifurcó en dos direcciones que definirían el futuro del procesamiento del lenguaje:

```mermaid
flowchart TB
    TRANS["Transformer 2017"] --> ENC["Solo Encoder"]
    TRANS --> DEC["Solo Decoder"]
    
    ENC --> BERT["BERT: Comprensión"]
    DEC --> GPT["GPT: Generación"]
    
    style BERT fill:#4285f4,color:#fff
    style GPT fill:#10a37f,color:#fff
```

**BERT** utilizó solo el encoder para crear representaciones bidireccionales, revolucionando tareas de comprensión. **GPT** tomó el camino opuesto: solo el decoder, generando texto de izquierda a derecha. Este enfoque **autoregresivo** resultó ser la base de los LLM actuales.

## La era de los Large Language Models

Con la arquitectura Transformer establecida, la investigación descubrió que escalar los modelos producía mejoras no lineales. Las **scaling laws** (leyes de escalado) formalizaron esta relación: aumentar parámetros, datos y cómputo de forma coordinada produce **capacidades emergentes** que no estaban presentes en modelos más pequeños.

### Difusión y convergencia hacia Transformers

A partir de 2020, los **modelos de difusión** ganaron protagonismo en la generación de imágenes. Su proceso consiste en añadir ruido gradualmente a una imagen hasta convertirla en ruido puro, y luego entrenar al modelo para revertir este proceso. Los primeros modelos usaban arquitecturas **U-Net** para este proceso de eliminación de ruido.

Sin embargo, en 2023-2024 se produjo una evolución significativa: los **Diffusion Transformers** (DiT) reemplazaron la U-Net por arquitecturas Transformer. Esta convergencia es visible en los modelos más avanzados de generación de imágenes y video.

> Prácticamente todas las modalidades (texto, imagen, audio, video) han convergido hacia arquitecturas basadas en Transformers, consolidándolo como la arquitectura dominante del estado del arte actual.

### Alineación con preferencias humanas

El salto de modelos capaces a modelos *útiles* requirió una técnica nueva: **RLHF** (Reinforcement Learning from Human Feedback). En lugar de solo predecir la siguiente palabra, los modelos aprenden a generar respuestas que los humanos prefieren.

El proceso funciona en tres fases:

* **Pre-entrenamiento**: el modelo aprende patrones del lenguaje con grandes cantidades de texto.
* **Instruction tuning**: se ajusta con ejemplos de instrucciones y respuestas deseadas.
* **RLHF**: evaluadores humanos puntúan respuestas, y el modelo se optimiza para maximizar esas puntuaciones.

> RLHF transformó los LLM de predictores de texto en asistentes que comprenden intenciones, siguen instrucciones y evitan contenido dañino.

### Arquitecturas eficientes

A medida que los modelos crecían, surgió la necesidad de ejecutarlos de forma más eficiente. Dos técnicas destacan:

* **Mixture of Experts (MoE)**: en lugar de activar todos los parámetros para cada token, el modelo selecciona dinámicamente un subconjunto de "expertos" especializados. Un modelo puede tener cientos de miles de millones de parámetros totales pero activar solo una fracción en cada inferencia, reduciendo el coste computacional.

* **Quantization**: representar los pesos del modelo con menor precisión numérica (8 bits o 4 bits en lugar de 16 o 32) permite ejecutar modelos grandes en hardware más accesible sin pérdida significativa de calidad.

### Tendencias actuales

El estado del arte evoluciona en varias direcciones simultáneas:

* **Razonamiento configurable**: los modelos actuales permiten ajustar el nivel de razonamiento mediante parámetros. Por ejemplo, se puede elegir entre modos como *instant* (respuestas rápidas), *thinking* (análisis profundo) o niveles graduales de esfuerzo de razonamiento, adaptando el coste computacional a la complejidad de cada tarea.

* **Multimodalidad nativa**: modelos entrenados desde el inicio para procesar texto, imágenes, audio y video de forma integrada. Las arquitecturas actuales unifican todas las modalidades en una misma secuencia, eliminando la necesidad de módulos separados.

* **Generación de video**: modelos como Sora, Veo y Movie Gen generan videos realistas a partir de descripciones textuales, aplicando arquitecturas Transformer a la generación temporal con audio sincronizado.

* **Open-weight**: la disponibilidad de modelos con pesos públicos (LLaMA, Qwen, Mistral, Gemma) ha democratizado el acceso y acelerado la innovación, permitiendo ejecución local y personalización.

### Direcciones futuras

La investigación explora varias líneas que podrían definir la próxima generación de modelos:

* **Razonamiento en espacios latentes**: en lugar de generar cadenas de pensamiento en texto visible, los modelos podrían razonar directamente en representaciones internas compactas, reduciendo el coste computacional y mejorando la coherencia.

* **Test-time compute scaling**: escalar el cómputo durante la inferencia (no solo en entrenamiento) para que los modelos dediquen más recursos a problemas difíciles, adaptando dinámicamente su esfuerzo.

* **State Space Models (Mamba, RWKV)**: arquitecturas con complejidad lineal en lugar de cuadrática, prometiendo mayor eficiencia en secuencias largas. Aún no han desplazado a los Transformers en benchmarks principales.

* **World Models**: propuestos por investigadores como Yann LeCun, buscan construir representaciones internas del mundo que permitan razonar y planificar, en lugar de solo predecir la siguiente palabra.

> El debate sobre el futuro de las arquitecturas permanece abierto: mientras los Transformers siguen mejorando con técnicas como razonamiento latente y escalado de inferencia, la investigación explora alternativas que podrían redefinir el campo.
