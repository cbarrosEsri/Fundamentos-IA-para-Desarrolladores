Las siguientes secciones explican cada rol en detalle y cómo se relacionan entre sí.

## Un nuevo ecosistema profesional

La adopción masiva de la IA Generativa ha transformado el panorama laboral tecnológico. Los roles tradicionales de Machine Learning se han especializado y han surgido disciplinas completamente nuevas que no existían hace pocos años.

Esta evolución responde a una realidad práctica: trabajar con LLMs requiere habilidades diferentes a las del Machine Learning clásico. No se trata tanto de entrenar modelos desde cero como de **integrar, orquestar y optimizar** modelos existentes para resolver problemas reales.

### El mercado laboral actual

Es importante distinguir entre la realidad del mercado y las expectativas:

* **Roles consolidados**: Data Engineer, ML Engineer, Data Scientist siguen siendo los más demandados en ofertas de empleo tradicionales.

* **Roles emergentes**: AI Engineer, LLMOps aparecen cada vez más, especialmente en empresas tecnológicas y startups.

* **Roles especializados**: Agent Engineer, Context Engineer son términos que se usan más en la industria que en ofertas formales de empleo, pero representan funciones reales dentro de equipos de desarrollo.

> Muchas empresas buscan "desarrolladores con experiencia en IA" o "ingenieros de software con conocimientos de LLMs" en lugar de títulos específicos. Las habilidades importan más que el nombre del puesto.

El cambio fundamental es que ahora se puede construir aplicaciones de IA sin entrenar modelos propios. Esto ha democratizado el acceso pero también ha creado la necesidad de nuevas especializaciones.

## Prompt Engineering

El **Prompt Engineering** es la disciplina de diseñar instrucciones efectivas para obtener resultados óptimos de los modelos de lenguaje. Aunque pueda parecer simple, formular prompts que produzcan respuestas consistentes, precisas y útiles requiere conocimiento técnico y metodología.

### Responsabilidades del Prompt Engineer

Un Prompt Engineer se encarga de:

* Diseñar **plantillas de prompts** reutilizables para diferentes casos de uso.
* Experimentar con diferentes formulaciones para optimizar la calidad de las respuestas.
* Documentar patrones efectivos y crear guías de estilo para el equipo.
* Evaluar sistemáticamente la calidad de las salidas mediante métricas y revisión humana.
* Adaptar prompts cuando cambian los modelos o requisitos del negocio.

### Técnicas fundamentales

El Prompt Engineering incluye técnicas como:

* **Zero-shot prompting**: pedir al modelo que realice una tarea sin ejemplos previos.
* **Few-shot prompting**: proporcionar ejemplos de entrada-salida esperada antes de la tarea real.
* **Chain of Thought**: solicitar que el modelo razone paso a paso antes de dar la respuesta final.
* **Role prompting**: asignar un rol o persona al modelo para guiar su comportamiento.

```mermaid
flowchart LR
    subgraph Tecnicas["Técnicas de prompting"]
        ZS[Zero-shot]
        FS[Few-shot]
        COT[Chain of Thought]
        RP[Role prompting]
    end
    
    ZS --> RES[Respuesta<br/>optimizada]
    FS --> RES
    COT --> RES
    RP --> RES
    
    style RES fill:#9f7aea,color:#fff
```

> El Prompt Engineering es la puerta de entrada al trabajo con IA Generativa. No requiere saber programar, pero sí entender cómo funcionan los modelos y pensar de forma estructurada.

### De Prompt Engineering a Context Engineering

El Prompt Engineering ha evolucionado hacia una disciplina más amplia: el **Context Engineering**. Mientras que el Prompt Engineering se centra en escribir instrucciones efectivas, el Context Engineering aborda la pregunta más amplia de *qué configuración de contexto es más probable que genere el comportamiento deseado del modelo*.

El contexto incluye mucho más que el prompt:

* **System prompts**: las instrucciones base del modelo.
* **Herramientas disponibles**: definiciones de tools que el modelo puede invocar.
* **Datos externos**: información recuperada mediante RAG u otras fuentes.
* **Historial de mensajes**: conversaciones previas y resultados de herramientas.
* **Protocolos de contexto**: estándares como MCP (Model Context Protocol) para conectar modelos con fuentes de datos.

```mermaid
flowchart LR
    subgraph PE["Prompt Engineering"]
        P[System prompt]
    end
    
    subgraph CE["Context Engineering"]
        SP[System prompt]
        T[Tools]
        D[Datos externos]
        H[Historial]
        MCP[MCP]
    end
    
    PE --> CE
    
    style CE fill:#9f7aea,color:#fff
```

> El Context Engineering trata el contexto como un recurso finito con rendimientos decrecientes. A medida que aumenta el número de tokens, la capacidad del modelo para procesar información disminuye. El arte está en encontrar el conjunto mínimo de tokens de alta señal.

Esta evolución es especialmente relevante para agentes que operan en múltiples turnos de inferencia, donde el contexto debe refinarse continuamente para mantener la coherencia sin saturar la ventana de contexto.

## AI Engineering

El **AI Engineering** es la disciplina de construir aplicaciones completas que integran modelos de IA Generativa. A diferencia del ML Engineer tradicional que entrena modelos, el AI Engineer se centra en **orquestar componentes** para crear productos funcionales.

### Diferencia con ML Engineering

| Aspecto | ML Engineering | AI Engineering |
|---------|----------------|----------------|
| Foco principal | Entrenar y optimizar modelos | Integrar modelos en aplicaciones |
| Datos | Preparar datasets de entrenamiento | Gestionar contexto y prompts |
| Infraestructura | Clusters de entrenamiento | APIs y orquestación |
| Output | Modelo entrenado | Aplicación funcional |

### Stack tecnológico

Un AI Engineer trabaja con un conjunto de tecnologías específico:

* **Frameworks de orquestación**: LangChain, LlamaIndex para conectar modelos con datos y herramientas.
* **APIs de proveedores**: OpenAI, Anthropic, Google para acceder a modelos.
* **Bases de datos vectoriales**: para implementar búsqueda semántica y RAG.
* **Model Context Protocol (MCP)**: estándar abierto para conectar modelos con fuentes de datos y herramientas de forma estandarizada.
* **Herramientas de evaluación**: para medir calidad de respuestas y detectar regresiones.

```mermaid
flowchart TB
    AIE[AI Engineer] --> APP[Aplicación GenAI]
    
    subgraph Componentes["Componentes que integra"]
        LLM[APIs de LLMs]
        VDB[Base de datos vectorial]
        TOOLS[Herramientas externas]
        UI[Interfaz de usuario]
    end
    
    LLM --> APP
    VDB --> APP
    TOOLS --> APP
    UI --> APP
    
    style AIE fill:#9f7aea,color:#fff
```

### Responsabilidades típicas

El día a día de un AI Engineer incluye:

* Diseñar la **arquitectura de la aplicación** combinando diferentes componentes.
* Implementar pipelines de RAG para conectar modelos con datos actualizados.
* Gestionar el **contexto y la memoria** en conversaciones multi-turno.
* Integrar herramientas externas que el modelo pueda invocar.
* Optimizar latencia y costes de las llamadas a APIs.

> El AI Engineer es el constructor de aplicaciones de IA Generativa. Necesita entender tanto el funcionamiento de los modelos como los principios de ingeniería de software tradicional.

## Agent Engineering

El **Agent Engineering** es la disciplina más reciente, centrada en diseñar y desarrollar **agentes autónomos** capaces de ejecutar tareas complejas con mínima supervisión humana.

### Qué diferencia a un agente

Un agente no es simplemente un chatbot. Los agentes pueden:

* **Planificar**: descomponer objetivos complejos en pasos manejables.
* **Ejecutar acciones**: invocar herramientas, APIs y sistemas externos.
* **Razonar sobre resultados**: evaluar si una acción tuvo éxito y decidir los siguientes pasos.
* **Iterar**: ajustar su enfoque cuando algo no funciona como esperaba.

```mermaid
flowchart LR
    U[Usuario] --> O[Objetivo]
    O --> AG[Agente]
    
    subgraph Ciclo["Ciclo del agente"]
        P[Planificar] --> E[Ejecutar]
        E --> EV[Evaluar]
        EV --> P
    end
    
    AG --> Ciclo
    Ciclo --> R[Resultado]
    
    style AG fill:#9f7aea,color:#fff
```

### Responsabilidades del Agent Engineer

Un Agent Engineer se encarga de:

* Diseñar la **arquitectura del agente**: qué herramientas tiene disponibles, cómo decide cuál usar, cómo gestiona errores.
* Definir **guardrails**: límites de lo que el agente puede y no puede hacer para evitar comportamientos no deseados.
* Implementar mecanismos de **supervisión humana**: puntos donde el agente solicita aprobación antes de ejecutar acciones críticas.
* Optimizar el **razonamiento**: mejorar cómo el agente descompone problemas y toma decisiones.
* Gestionar el **estado y memoria**: mantener contexto relevante a través de múltiples interacciones.

### Context Engineering para agentes

Los agentes que operan durante períodos prolongados requieren técnicas específicas de gestión de contexto:

* **Compactación**: resumir el historial de conversación cuando se acerca al límite de la ventana de contexto, preservando decisiones clave y descartando información redundante.

* **Notas estructuradas**: el agente mantiene notas persistentes fuera de la ventana de contexto (como un archivo de to-do o notas de progreso) que puede consultar en turnos posteriores.

* **Arquitecturas multi-agente**: dividir tareas complejas entre sub-agentes especializados que operan con contextos limpios y devuelven resúmenes condensados al agente principal.

> El Agent Engineering requiere pensar en sistemas autónomos que operan con incertidumbre. Es una mezcla de diseño de sistemas, ingeniería de control y comprensión profunda de las capacidades y limitaciones de los LLMs.

## LLMOps

**LLMOps** (Large Language Model Operations) adapta los principios de MLOps al contexto específico de los modelos de lenguaje. Se centra en llevar aplicaciones de IA Generativa a **producción de forma fiable y escalable**.

### Diferencias con MLOps tradicional

| Aspecto | MLOps | LLMOps |
|---------|-------|--------|
| Entrenamiento | Frecuente, con datos propios | Poco frecuente, se usan modelos existentes |
| Evaluación | Métricas numéricas claras | Evaluación cualitativa compleja |
| Costes | Principalmente entrenamiento | Principalmente inferencia |
| Versionado | Pesos del modelo | Prompts, configuraciones, pipelines |

### Áreas de responsabilidad

Un LLMOps Engineer gestiona:

* **Despliegue y escalado**: configurar infraestructura para servir modelos con baja latencia y alta disponibilidad.
* **Monitorización**: detectar degradación de calidad, alucinaciones frecuentes o cambios en el comportamiento del modelo.
* **Gestión de costes**: optimizar uso de tokens, implementar caching, elegir modelos apropiados para cada tarea.
* **Versionado de prompts**: mantener control de cambios en plantillas y configuraciones.
* **Evaluación continua**: implementar pipelines de testing para detectar regresiones cuando cambian modelos o prompts.

```mermaid
flowchart TB
    subgraph LLMOps["Responsabilidades LLMOps"]
        DEP[Despliegue]
        MON[Monitorización]
        COST[Gestión costes]
        VER[Versionado]
        EVAL[Evaluación]
    end
    
    DEP --> PROD[Producción<br/>fiable]
    MON --> PROD
    COST --> PROD
    VER --> PROD
    EVAL --> PROD
    
    style PROD fill:#9f7aea,color:#fff
```

> LLMOps es esencial cuando las aplicaciones de IA Generativa pasan de prototipos a sistemas en producción con usuarios reales y requisitos de disponibilidad.

## El desarrollador aumentado por IA

Más allá de los roles especializados, la IA Generativa está transformando el trabajo de **todos los desarrolladores de software**. Herramientas como asistentes de código integrados en IDEs han creado una nueva forma de trabajar.

### Impacto en la productividad

Los desarrolladores que utilizan asistentes de IA reportan:

* Reducción significativa del tiempo en tareas repetitivas como escribir boilerplate o tests.
* Mayor velocidad en la exploración de codebases desconocidas.
* Generación automática de documentación y comentarios.
* Asistencia en debugging y refactorización.

### Nuevas habilidades necesarias

Esta transformación no elimina la necesidad de programadores, pero cambia las habilidades más valoradas:

* **Capacidad de revisión crítica**: evaluar y corregir código generado por IA.
* **Especificación precisa**: saber explicar qué se quiere construir de forma clara.
* **Arquitectura y diseño**: las decisiones de alto nivel siguen siendo humanas.
* **Integración y contexto**: entender cómo encaja cada pieza en el sistema completo.

> La IA generativa amplifica las capacidades de los desarrolladores experimentados más que las de los principiantes. Saber qué preguntar y cómo validar las respuestas requiere conocimiento previo del dominio.

## Roles complementarios

Además de los roles técnicos principales, el ecosistema de IA Generativa incluye otras especializaciones:

* **AI Ethics Specialist**: evalúa sesgos, garantiza cumplimiento normativo y define políticas de uso responsable.

* **AI Product Manager**: traduce necesidades de negocio en requisitos técnicos y prioriza el desarrollo de funcionalidades de IA.

* **Data Engineer para GenAI**: prepara y mantiene los datos que alimentan sistemas RAG y bases de conocimiento.

* **AI Trainer / Evaluator**: genera datos de evaluación, califica respuestas y proporciona feedback para mejorar los modelos.

* **AI Security Specialist**: se especializa en proteger sistemas de IA contra prompt injection, jailbreaks y otras vulnerabilidades específicas de LLMs.

### Trayectorias profesionales

Las fronteras entre roles son fluidas y muchos profesionales combinan habilidades de varias disciplinas. Una trayectoria típica podría ser:

* Comenzar con **Prompt Engineering** para entender cómo funcionan los modelos.
* Evolucionar hacia **Context Engineering** dominando la gestión integral del contexto.
* Avanzar hacia **AI Engineering** añadiendo habilidades de desarrollo e integración.
* Especializarse en **Agent Engineering** o **LLMOps** según intereses.

La clave está en construir una base sólida de comprensión de los modelos y complementarla con habilidades técnicas específicas según el rol deseado.
