## El sistema de mensajes en LangChain

Los modelos de chat modernos no procesan texto plano, sino **secuencias de mensajes** con roles diferenciados. Cada mensaje tiene un rol que indica quién lo envió: el sistema, el usuario o el asistente. Esta estructura permite al modelo entender el contexto de la conversación.

LangChain proporciona tres clases principales para representar estos mensajes:

* **SystemMessage** - Instrucciones para configurar el comportamiento del modelo
* **HumanMessage** - Mensajes enviados por el usuario
* **AIMessage** - Respuestas generadas por el modelo

> El orden de los mensajes es importante. Típicamente, una conversación comienza con un SystemMessage opcional, seguido de una alternancia entre HumanMessage y AIMessage.

Para utilizar estos tipos de mensajes, se importan desde `langchain.messages`:

````python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
````

## Formatos de mensajes

LangChain ofrece varias formas de especificar mensajes. La más común es usar **tuplas** con el formato `(rol, contenido)`:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

mensajes = [
    ("system", "Eres un asistente experto en historia española. Responde de forma concisa."),
    ("human", "¿Quién fue el primer rey de España?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

Los roles disponibles son:

* `"system"` - Equivalente a SystemMessage
* `"human"` o `"user"` - Equivalente a HumanMessage
* `"assistant"` o `"ai"` - Equivalente a AIMessage

### Formato con objetos Message

También puedes usar los objetos Message explícitamente, lo cual es útil cuando necesitas acceder a propiedades adicionales:

````python
from langchain.messages import SystemMessage, HumanMessage
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

mensajes = [
    SystemMessage(content="Eres un asistente experto en historia española."),
    HumanMessage(content="¿Quién fue el primer rey de España?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

> Ambos formatos son válidos y producen el mismo resultado. El formato de tuplas es más conciso, mientras que los objetos Message permiten añadir metadatos adicionales.

### Formato diccionario

LangChain también acepta mensajes en formato diccionario, compatible con el formato de OpenAI:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

mensajes = [
    {"role": "system", "content": "Eres un traductor al inglés."},
    {"role": "user", "content": "Buenos días"}
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

## SystemMessage: definir el comportamiento

El **SystemMessage** establece las instrucciones base que guían el comportamiento del modelo durante toda la conversación. Es como dar un rol o personalidad al asistente antes de que comience a interactuar.

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

mensajes = [
    ("system", "Eres un asistente experto en historia española. Responde de forma concisa."),
    ("human", "¿Quién fue el primer rey de España?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

El SystemMessage no aparece en la respuesta, pero **condiciona** cómo el modelo procesa y responde a los mensajes posteriores. En este ejemplo, el modelo responderá desde la perspectiva de un experto en historia española y de forma concisa.

### Usos comunes del SystemMessage

El SystemMessage es útil para varios propósitos:

* **Definir un rol específico**: "Eres un profesor de matemáticas"
* **Establecer el formato de respuesta**: "Responde siempre en formato JSON"
* **Limitar el alcance**: "Solo responde preguntas sobre programación en Python"
* **Definir el tono**: "Usa un tono formal y profesional"

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

# Sistema que traduce al inglés
mensajes = [
    ("system", "Eres un traductor. Traduce todo lo que te digan al inglés."),
    ("human", "Buenos días, ¿cómo estás?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
# Good morning, how are you?
````

> El SystemMessage es opcional pero muy recomendable. Sin él, el modelo actúa como un asistente genérico sin instrucciones específicas.

## HumanMessage: mensajes del usuario

El **HumanMessage** representa los mensajes que envía el usuario al modelo. Es el tipo de mensaje más común y el que inicia las interacciones.

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

# Invocar con un solo mensaje del usuario
mensajes = [
    ("human", "¿Cuántos planetas tiene el sistema solar?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

Cuando se invoca el modelo con una cadena de texto directamente, LangChain internamente lo convierte en un HumanMessage. Las siguientes llamadas son equivalentes:

````python
from langchain.messages import HumanMessage
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

# Forma abreviada (cadena de texto)
respuesta1 = modelo.invoke("Hola")

# Forma con tupla
respuesta2 = modelo.invoke([("human", "Hola")])

# Forma explícita con objeto HumanMessage
respuesta3 = modelo.invoke([HumanMessage(content="Hola")])
````

## AIMessage: respuestas del modelo

El **AIMessage** representa las respuestas generadas por el modelo. Cuando invocas un modelo, el resultado que obtienes es un objeto AIMessage.

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

respuesta = modelo.invoke("¿Qué es Python?")

# La respuesta es un AIMessage
print(type(respuesta))
# <class 'langchain.messages.ai.AIMessage'>

print(respuesta.content)
````

Además de recibirlos como respuesta, también puedes incluir AIMessage en el historial para simular respuestas previas del modelo:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

mensajes = [
    ("system", "Eres un asistente de cocina."),
    ("human", "¿Qué ingredientes necesito para una tortilla?"),
    ("assistant", "Para una tortilla española necesitas: huevos, patatas, aceite de oliva y sal. Opcionalmente, cebolla."),
    ("human", "¿Cuántos huevos para 4 personas?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

En este ejemplo, el mensaje con rol `"assistant"` proporciona contexto sobre una respuesta anterior, permitiendo que la siguiente pregunta tenga sentido.

## Construir conversaciones acumulando mensajes

Para mantener una **conversación con contexto**, es necesario acumular todos los mensajes previos y enviarlos junto con cada nueva pregunta. El modelo no recuerda interacciones anteriores por sí mismo.

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

# Iniciar la conversación con un sistema
historial = [
    ("system", "Eres un asistente amable que ayuda con dudas de programación.")
]

# Primera pregunta
historial.append(("human", "¿Qué es una variable en Python?"))
respuesta = modelo.invoke(historial)
historial.append(("assistant", respuesta.content))
print(f"Asistente: {respuesta.content}\n")

# Segunda pregunta (el modelo tiene contexto de la anterior)
historial.append(("human", "¿Y cómo le asigno un valor?"))
respuesta = modelo.invoke(historial)
historial.append(("assistant", respuesta.content))
print(f"Asistente: {respuesta.content}\n")

# Tercera pregunta
historial.append(("human", "Dame un ejemplo con números"))
respuesta = modelo.invoke(historial)
historial.append(("assistant", respuesta.content))
print(f"Asistente: {respuesta.content}")
````

En cada invocación, el modelo recibe **todo el historial** de mensajes. Esto le permite entender el contexto y responder de forma coherente a preguntas que hacen referencia a turnos anteriores.

> Cada vez que se invoca el modelo, se envía el historial completo. Esto significa que conversaciones largas consumirán más tokens y tiempo de procesamiento.

### Patrón de conversación interactiva

Un patrón común es crear un bucle que permita al usuario interactuar de forma continua:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

historial = [
    ("system", "Eres un asistente conversacional. Responde de forma breve y clara.")
]

print("Escribe 'salir' para terminar la conversación.\n")

while True:
    # Obtener entrada del usuario
    entrada = input("Tú: ")
    
    if entrada.lower() == "salir":
        print("Conversación finalizada.")
        break
    
    # Añadir mensaje del usuario al historial
    historial.append(("human", entrada))
    
    # Obtener respuesta del modelo
    respuesta = modelo.invoke(historial)
    
    # Añadir respuesta al historial
    historial.append(("assistant", respuesta.content))
    
    print(f"Asistente: {respuesta.content}\n")
````

Este código crea una conversación interactiva donde cada mensaje se acumula en el historial. El modelo siempre tiene acceso a toda la conversación previa para generar respuestas contextualizadas.

### Inspeccionar el historial

Puedes examinar el contenido del historial en cualquier momento para depurar o entender el flujo de la conversación:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

historial = [
    ("system", "Eres un asistente de viajes.")
]

historial.append(("human", "Quiero ir a Barcelona"))
respuesta = modelo.invoke(historial)
historial.append(("assistant", respuesta.content))

historial.append(("human", "¿Qué puedo visitar allí?"))
respuesta = modelo.invoke(historial)
historial.append(("assistant", respuesta.content))

# Inspeccionar el historial completo
for i, mensaje in enumerate(historial):
    rol, contenido = mensaje
    contenido_corto = contenido[:50] + "..." if len(contenido) > 50 else contenido
    print(f"{i}. [{rol}] {contenido_corto}")
````

La salida mostraría algo como:

```
0. [system] Eres un asistente de viajes.
1. [human] Quiero ir a Barcelona
2. [assistant] Barcelona es una ciudad maravillosa...
3. [human] ¿Qué puedo visitar allí?
4. [assistant] En Barcelona puedes visitar la Sagrada Familia...
```

Esta técnica es útil para verificar que los mensajes se están acumulando correctamente y para entender qué información tiene disponible el modelo en cada turno de la conversación.
