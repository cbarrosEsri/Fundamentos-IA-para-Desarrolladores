## Obtener una API key de OpenAI

Para utilizar los modelos de OpenAI desde LangChain, necesitas una **API key** que autentique tus peticiones. Este proceso se realiza desde la plataforma de desarrolladores de OpenAI.

El primer paso es crear una cuenta en [platform.openai.com](https://platform.openai.com). Si ya tienes una cuenta de ChatGPT, puedes usar las mismas credenciales. Una vez dentro, navega a la sección **API keys** en el menú lateral.


Para generar una nueva clave, haz clic en **Create new secret key**. Opcionalmente puedes asignarle un nombre descriptivo como "langchain-curso". Al confirmar, OpenAI mostrará la clave completa **una única vez**, así que cópiala inmediatamente.

> La API key es confidencial y no debe compartirse ni subirse a repositorios públicos. Trátala como una contraseña.

### Configurar facturación

OpenAI requiere un método de pago válido para usar la API. Desde la sección **Billing** puedes añadir una tarjeta de crédito y establecer límites de gasto mensuales. Los modelos tienen precios diferentes según su capacidad:

* **GPT-5-mini**: El más económico, ideal para desarrollo y pruebas
* **GPT-4.1**: Balance entre coste y rendimiento
* **GPT-5**: Mayor capacidad, coste intermedio
* **GPT-5.1 y GPT-5.2**: Máximo rendimiento, coste más elevado

Para comenzar, un presupuesto de 5-10 USD es suficiente para experimentar con todos los ejemplos de este curso.

## Configurar la API key con archivo .env

LangChain carga automáticamente las variables de entorno desde un archivo `.env` ubicado en el directorio de trabajo. Esta es la forma recomendada de gestionar credenciales sin exponerlas en el código.

Crea un archivo llamado `.env` en la raíz de tu proyecto:

```
OPENAI_API_KEY=sk-proj-tu-clave-aqui
```

Sustituye `sk-proj-tu-clave-aqui` por la API key que copiaste anteriormente. El prefijo `sk-proj-` es el formato estándar de las claves de proyecto de OpenAI.

Con esta configuración, **ChatOpenAI detectará automáticamente la variable** `OPENAI_API_KEY` y la utilizará para autenticar las peticiones. No es necesario cargar manualmente el archivo ni usar bibliotecas adicionales como `python-dotenv`.

````python
from langchain_openai import ChatOpenAI

# La API key se carga automáticamente desde .env
modelo = ChatOpenAI(model="gpt-5-mini")

respuesta = modelo.invoke("Hola, ¿cómo estás?")
print(respuesta.content)
````

> Añade el archivo `.env` a tu `.gitignore` para evitar subir credenciales a repositorios de código.

## Instanciar ChatOpenAI

La clase **ChatOpenAI** es el punto de entrada para interactuar con los modelos de OpenAI desde LangChain. Funciona de manera análoga a ChatOllama, pero conectándose a los servidores de OpenAI en lugar de a un servidor local.

````python
from langchain_openai import ChatOpenAI

# Instanciar con el modelo más económico
modelo = ChatOpenAI(model="gpt-5-mini")

# Invocar con texto simple
respuesta = modelo.invoke("¿Cuál es la capital de España?")
print(respuesta.content)
````

El parámetro `model` especifica qué modelo de OpenAI utilizar. Los modelos disponibles actualmente incluyen:

* **gpt-5.1**: El modelo más avanzado de OpenAI, máxima capacidad de razonamiento
* **gpt-5**: Excelente rendimiento para tareas complejas
* **gpt-5-mini**: Versión compacta y económica, ideal para desarrollo
* **gpt-4.1**: Balance óptimo entre coste y rendimiento

### Verificar la conexión

Antes de desarrollar aplicaciones complejas, es recomendable verificar que la configuración es correcta:

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-mini")

# Test simple de conexión
respuesta = modelo.invoke("Di 'hola'")
print(f"Respuesta: {respuesta.content}")
print(f"Modelo usado: {respuesta.response_metadata.get('model_name')}")
````

Si la API key es incorrecta o no está configurada, recibirás un error de autenticación con información sobre el problema.

## Invocar modelos y obtener respuestas

El método `invoke` funciona exactamente igual que con ChatOllama. Acepta texto, listas de mensajes o cualquiera de los formatos de entrada que vimos en lecciones anteriores.

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-mini")

# Invocación con texto simple
respuesta = modelo.invoke("Explica qué es Python en una frase")
print(respuesta.content)
````

La respuesta es un objeto **AIMessage** con la misma estructura que las respuestas de Ollama:

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-mini")
respuesta = modelo.invoke("¿Qué es una API?")

# Contenido de la respuesta
print(f"Respuesta: {respuesta.content}")

# Metadatos de uso
print(f"Tokens de entrada: {respuesta.usage_metadata.get('input_tokens')}")
print(f"Tokens de salida: {respuesta.usage_metadata.get('output_tokens')}")
print(f"Total tokens: {respuesta.usage_metadata.get('total_tokens')}")
````

> Los tokens consumidos determinan el coste de cada petición. Monitorizar el uso de tokens te ayuda a controlar los gastos.

### Usar diferentes modelos

Puedes cambiar entre modelos simplemente modificando el parámetro `model`. Cada modelo tiene características diferentes:

````python
from langchain_openai import ChatOpenAI

pregunta = "Explica la teoría de la relatividad en términos simples"

# Modelo económico para pruebas
modelo_mini = ChatOpenAI(model="gpt-5-mini")
print("=== GPT-5-mini ===")
print(modelo_mini.invoke(pregunta).content)

# Modelo avanzado para tareas complejas
modelo_avanzado = ChatOpenAI(model="gpt-5")
print("\n=== GPT-5 ===")
print(modelo_avanzado.invoke(pregunta).content)
````

### Conversaciones con mensajes estructurados

Los modelos de OpenAI soportan todos los formatos de mensajes que vimos anteriormente:

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-mini")

# Usando tuplas con roles
mensajes = [
    ("system", "Eres un experto en historia del arte. Responde de forma concisa."),
    ("human", "¿Quién pintó la Mona Lisa?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

También puedes usar el formato diccionario compatible con la API nativa de OpenAI:

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-mini")

mensajes = [
    {"role": "system", "content": "Eres un traductor al inglés."},
    {"role": "user", "content": "Buenos días, ¿cómo estás?"}
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

## Comparar OpenAI con Ollama

Una ventaja de usar LangChain es que la **interfaz es idéntica** para diferentes proveedores. Esto permite cambiar entre modelos locales y en la nube con cambios mínimos en el código.

````python
from langchain_openai import ChatOpenAI
from langchain_ollama import ChatOllama

pregunta = "¿Cuántos planetas tiene el sistema solar?"

# Modelo en la nube (OpenAI)
modelo_openai = ChatOpenAI(model="gpt-5-mini")
print("=== OpenAI GPT-5-mini ===")
respuesta_openai = modelo_openai.invoke(pregunta)
print(respuesta_openai.content)
print(f"Tokens: {respuesta_openai.usage_metadata.get('total_tokens')}")

# Modelo local (Ollama)
modelo_ollama = ChatOllama(model="gemma3:1b")
print("\n=== Ollama Gemma3 ===")
respuesta_ollama = modelo_ollama.invoke(pregunta)
print(respuesta_ollama.content)
print(f"Tokens: {respuesta_ollama.usage_metadata.get('total_tokens')}")
````
