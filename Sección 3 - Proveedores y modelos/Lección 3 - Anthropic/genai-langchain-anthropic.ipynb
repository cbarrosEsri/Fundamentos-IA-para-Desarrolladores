## Anthropic y la familia de modelos Claude

Anthropic es una empresa de investigación en seguridad de inteligencia artificial fundada en 2021. Su producto principal es **Claude**, una familia de modelos de lenguaje que se distingue por su enfoque en seguridad, capacidad de razonamiento y contexto extenso.

La familia Claude incluye tres niveles de modelos diseñados para diferentes casos de uso:

* **Claude Haiku**: El modelo más rápido y económico, ideal para tareas de alto volumen
* **Claude Sonnet**: Balance óptimo entre rendimiento y coste, adecuado para la mayoría de aplicaciones
* **Claude Opus**: Máxima capacidad de razonamiento para tareas complejas

## Obtener una API key de Anthropic

Para utilizar los modelos de Anthropic desde LangChain, necesitas una **API key** que autentique tus peticiones. Este proceso se realiza desde la consola de Anthropic.

El primer paso es crear una cuenta en [console.anthropic.com](https://console.anthropic.com). Una vez dentro, navega a la sección **API Keys** en el menú lateral.

```mermaid
flowchart LR
    subgraph Proceso["Obtener API key"]
        A["Crear cuenta"]
        B["Acceder a API Keys"]
        C["Crear nueva key"]
        D["Copiar y guardar"]
    end
    
    A --> B --> C --> D
    
    style A fill:#cc785c,color:#fff
    style B fill:#cc785c,color:#fff
    style C fill:#9b4d3b,color:#fff
    style D fill:#9b4d3b,color:#fff
```

Para generar una nueva clave, haz clic en **Create Key**. Puedes asignarle un nombre descriptivo como "langchain-curso". Al confirmar, Anthropic mostrará la clave completa **una única vez**, así que cópiala inmediatamente.

> La API key es confidencial y no debe compartirse ni subirse a repositorios públicos. Trátala como una contraseña.

### Configurar facturación

Anthropic requiere un método de pago válido para usar la API. Desde la sección **Billing** puedes añadir una tarjeta de crédito. Los modelos tienen precios diferentes según su capacidad, medidos en coste por millón de tokens de entrada y salida.

Para comenzar, un presupuesto de 5-10 USD es suficiente para experimentar con todos los ejemplos de este curso.

## Configurar la API key con archivo .env

LangChain carga automáticamente las variables de entorno desde un archivo `.env` ubicado en el directorio de trabajo. Esta es la forma recomendada de gestionar credenciales sin exponerlas en el código.

Crea un archivo llamado `.env` en la raíz de tu proyecto:

```
ANTHROPIC_API_KEY=sk-ant-api03-tu-clave-aqui
```

Sustituye `sk-ant-api03-tu-clave-aqui` por la API key que copiaste anteriormente. El prefijo `sk-ant-api03-` es el formato estándar de las claves de Anthropic.

Con esta configuración, **ChatAnthropic detectará automáticamente la variable** `ANTHROPIC_API_KEY` y la utilizará para autenticar las peticiones. No es necesario cargar manualmente el archivo ni usar bibliotecas adicionales.

````python
from langchain_anthropic import ChatAnthropic

# La API key se carga automáticamente desde .env
modelo = ChatAnthropic(model="claude-haiku-4-5-20251001")

respuesta = modelo.invoke("Hola, ¿cómo estás?")
print(respuesta.content)
````

> Añade el archivo `.env` a tu `.gitignore` para evitar subir credenciales a repositorios de código.

## Instanciar ChatAnthropic

La clase **ChatAnthropic** es el punto de entrada para interactuar con los modelos de Anthropic desde LangChain. Funciona de manera análoga a ChatOllama y ChatOpenAI, pero conectándose a los servidores de Anthropic.

````python
from langchain_anthropic import ChatAnthropic

# Instanciar con el modelo más económico
modelo = ChatAnthropic(model="claude-haiku-4-5-20251001")

# Invocar con texto simple
respuesta = modelo.invoke("¿Cuál es la capital de España?")
print(respuesta.content)
````

El parámetro `model` especifica qué modelo de Claude utilizar. Los identificadores de modelo de Anthropic incluyen la versión y fecha para garantizar reproducibilidad.

### Nomenclatura de modelos Claude

Anthropic utiliza un sistema de nomenclatura que incluye el nombre del modelo y la fecha de lanzamiento:

```
claude-{nivel}-{version}-{fecha}
```

Por ejemplo:
* `claude-haiku-4-5-20251001` - Claude Haiku versión 4.5
* `claude-sonnet-4-5-20250929` - Claude Sonnet versión 4.5
* `claude-opus-4-5-20251101` - Claude Opus versión 4.5

## Invocar modelos y obtener respuestas

El método `invoke` funciona exactamente igual que con otros proveedores. Acepta texto, listas de mensajes o cualquiera de los formatos de entrada que vimos en lecciones anteriores.

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(model="claude-haiku-4-5-20251001")

# Invocación con texto simple
respuesta = modelo.invoke("Explica qué es Python en una frase")
print(respuesta.content)
````

La respuesta es un objeto **AIMessage** con la misma estructura que las respuestas de otros proveedores:

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(model="claude-haiku-4-5-20251001")
respuesta = modelo.invoke("¿Qué es una API?")

# Contenido de la respuesta
print(f"Respuesta: {respuesta.content}")

# Metadatos de uso
print(f"Tokens de entrada: {respuesta.usage_metadata.get('input_tokens')}")
print(f"Tokens de salida: {respuesta.usage_metadata.get('output_tokens')}")
print(f"Total tokens: {respuesta.usage_metadata.get('total_tokens')}")
````

> Los tokens consumidos determinan el coste de cada petición. Monitorizar el uso de tokens te ayuda a controlar los gastos.

### Usar diferentes modelos

Puedes cambiar entre modelos simplemente modificando el parámetro `model`. Cada modelo tiene características diferentes en términos de velocidad, capacidad y coste:

````python
from langchain_anthropic import ChatAnthropic

pregunta = "Explica la teoría de la relatividad en términos simples"

# Modelo económico para pruebas
modelo_haiku = ChatAnthropic(model="claude-haiku-4-5-20251001")
print("=== Claude Haiku ===")
print(modelo_haiku.invoke(pregunta).content)

# Modelo avanzado para tareas complejas
modelo_sonnet = ChatAnthropic(model="claude-sonnet-4-5-20250929")
print("\n=== Claude Sonnet ===")
print(modelo_sonnet.invoke(pregunta).content)
````

### Conversaciones con mensajes estructurados

Los modelos de Anthropic soportan todos los formatos de mensajes estándar de LangChain:

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(model="claude-haiku-4-5-20251001")

# Usando tuplas con roles
mensajes = [
    ("system", "Eres un experto en historia del arte. Responde de forma concisa."),
    ("human", "¿Quién pintó la Mona Lisa?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

También puedes usar el formato diccionario:

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(model="claude-haiku-4-5-20251001")

mensajes = [
    {"role": "system", "content": "Eres un traductor al inglés."},
    {"role": "user", "content": "Buenos días, ¿cómo estás?"}
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

## Contar tokens antes de enviar

Anthropic ofrece una API oficial para contar tokens, lo que permite estimar costes antes de realizar una petición. LangChain expone esta funcionalidad a través del método `get_num_tokens_from_messages`:

````python
from langchain_anthropic import ChatAnthropic
from langchain.messages import HumanMessage, SystemMessage

modelo = ChatAnthropic(model="claude-sonnet-4-5-20250929")

mensajes = [
    SystemMessage(content="Eres un científico"),
    HumanMessage(content="Hola, Claude")
]

# Contar tokens antes de enviar
num_tokens = modelo.get_num_tokens_from_messages(mensajes)
print(f"Tokens en el mensaje: {num_tokens}")
````

Esta funcionalidad es útil para:

* **Estimar costes** antes de ejecutar peticiones largas
* **Validar** que el mensaje no excede el límite del contexto
* **Optimizar** prompts para reducir tokens innecesarios

## Comparar Anthropic con otros proveedores

Una ventaja de usar LangChain es que la **interfaz es idéntica** para diferentes proveedores. Esto permite cambiar entre modelos locales y en la nube con cambios mínimos en el código.

````python
from langchain_anthropic import ChatAnthropic
from langchain_openai import ChatOpenAI
from langchain_ollama import ChatOllama

pregunta = "¿Cuántos planetas tiene el sistema solar?"

# Modelo de Anthropic
modelo_anthropic = ChatAnthropic(model="claude-haiku-4-5-20251001")
print("=== Anthropic Claude Haiku ===")
respuesta_anthropic = modelo_anthropic.invoke(pregunta)
print(respuesta_anthropic.content)
print(f"Tokens: {respuesta_anthropic.usage_metadata.get('total_tokens')}")

# Modelo de OpenAI
modelo_openai = ChatOpenAI(model="gpt-5-mini")
print("\n=== OpenAI GPT-5-mini ===")
respuesta_openai = modelo_openai.invoke(pregunta)
print(respuesta_openai.content)
print(f"Tokens: {respuesta_openai.usage_metadata.get('total_tokens')}")

# Modelo local (Ollama)
modelo_ollama = ChatOllama(model="gemma3:1b")
print("\n=== Ollama Gemma3 ===")
respuesta_ollama = modelo_ollama.invoke(pregunta)
print(respuesta_ollama.content)
print(f"Tokens: {respuesta_ollama.usage_metadata.get('total_tokens')}")
````

## Características distintivas de Claude

### Ventana de contexto extendida

Claude destaca por su capacidad de procesar contextos muy largos. Los modelos más recientes soportan hasta **200K tokens** de contexto de forma estándar, con opción de extender hasta **1 millón de tokens** en beta.

````python
from langchain_anthropic import ChatAnthropic

# Modelo con contexto extendido (beta)
modelo = ChatAnthropic(
    model="claude-sonnet-4-5-20250929",
    betas=["context-1m-2025-08-07"]
)

# Ahora puedes enviar documentos muy largos
documento_largo = "..." * 100000  # Simula un documento extenso
respuesta = modelo.invoke(f"Resume este documento: {documento_largo}")
````

### Extended Thinking

Algunos modelos Claude soportan **extended thinking**, una funcionalidad que permite ver el proceso de razonamiento paso a paso que lleva a la respuesta final:

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(
    model="claude-sonnet-4-5-20250929",
    max_tokens=5000,
    thinking={"type": "enabled", "budget_tokens": 2000}
)

respuesta = modelo.invoke("¿Cuál es la raíz cúbica de 50.653?")

# La respuesta incluye bloques de razonamiento
for bloque in respuesta.content_blocks:
    if bloque.get("type") == "reasoning":
        print("Razonamiento:", bloque.get("reasoning"))
    elif bloque.get("type") == "text":
        print("Respuesta:", bloque.get("text"))
````

> Extended thinking consume tokens adicionales pero proporciona transparencia sobre cómo Claude llega a sus conclusiones. Es especialmente útil para tareas de razonamiento complejo.

### Esfuerzo de respuesta

Claude permite controlar el **esfuerzo** que dedica a generar una respuesta, balanceando calidad contra latencia y coste:

````python
from langchain_anthropic import ChatAnthropic

# Respuesta rápida con menos elaboración
modelo_rapido = ChatAnthropic(
    model="claude-opus-4-5-20251101",
    effort="low"
)

# Respuesta con máxima elaboración
modelo_detallado = ChatAnthropic(
    model="claude-opus-4-5-20251101",
    effort="high"
)

pregunta = "Analiza los pros y contras de la energía nuclear"

print("=== Esfuerzo bajo ===")
print(modelo_rapido.invoke(pregunta).content)

print("\n=== Esfuerzo alto ===")
print(modelo_detallado.invoke(pregunta).content)
````
