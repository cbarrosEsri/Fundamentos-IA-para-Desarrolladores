## Google y la familia de modelos Gemini

Google ofrece acceso a sus modelos de inteligencia artificial generativa a través de la familia **Gemini**. Estos modelos destacan por su capacidad multimodal nativa, pudiendo procesar texto, imágenes, audio, video y PDFs en una misma conversación.

La familia Gemini incluye diferentes variantes optimizadas para distintos casos de uso:

* **Gemini Pro**: El modelo más potente, diseñado para tareas complejas de razonamiento
* **Gemini Flash**: Balance óptimo entre rendimiento y velocidad, ideal para la mayoría de aplicaciones
* **Gemini Flash Lite**: Versión optimizada para alta velocidad y bajo coste

## Obtener una API key de Google AI

Para utilizar los modelos de Google desde LangChain, necesitas una **API key** que autentique tus peticiones. La forma más rápida de obtenerla es a través de Google AI Studio.

El primer paso es acceder a [aistudio.google.com/api-keys](https://aistudio.google.com/api-keys) con tu cuenta de Google. Desde ahí puedes generar una nueva API key con un solo clic.

> La API key es confidencial y no debe compartirse ni subirse a repositorios públicos. Trátala como una contraseña.

## Configurar la API key con archivo .env

LangChain carga automáticamente las variables de entorno desde un archivo `.env` ubicado en el directorio de trabajo. Esta es la forma recomendada de gestionar credenciales sin exponerlas en el código.

Crea un archivo llamado `.env` en la raíz de tu proyecto:

```
GOOGLE_API_KEY=tu-clave-aqui
```

Sustituye `tu-clave-aqui` por la API key que copiaste de Google AI Studio.

Con esta configuración, **ChatGoogleGenerativeAI detectará automáticamente la variable** `GOOGLE_API_KEY` y la utilizará para autenticar las peticiones. Como alternativa, también reconoce `GEMINI_API_KEY`.

````python
from langchain_google_genai import ChatGoogleGenerativeAI

# La API key se carga automáticamente desde .env
modelo = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

respuesta = modelo.invoke("Hola, ¿cómo estás?")
print(respuesta.content)
````

> Añade el archivo `.env` a tu `.gitignore` para evitar subir credenciales a repositorios de código.

## Instanciar ChatGoogleGenerativeAI

La clase **ChatGoogleGenerativeAI** es el punto de entrada para interactuar con los modelos de Google desde LangChain. Funciona de manera análoga a ChatOllama y ChatAnthropic, pero conectándose a los servidores de Google AI.

````python
from langchain_google_genai import ChatGoogleGenerativeAI

# Instanciar con el modelo flash (balance rendimiento/velocidad)
modelo = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# Invocar con texto simple
respuesta = modelo.invoke("¿Cuál es la capital de España?")
print(respuesta.content)
````

El parámetro `model` especifica qué modelo de Gemini utilizar. Los identificadores de modelo de Google siguen un patrón que incluye la familia y versión.

* `gemini-2.5-flash` - Gemini 2.5 Flash (recomendado para empezar)

## Invocar modelos y obtener respuestas

El método `invoke` funciona exactamente igual que con otros proveedores. Acepta texto, listas de mensajes o cualquiera de los formatos de entrada que vimos en lecciones anteriores.

````python
from langchain_google_genai import ChatGoogleGenerativeAI

modelo = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# Invocación con texto simple
respuesta = modelo.invoke("Explica qué es Python en una frase")
print(respuesta.content)
````

La respuesta es un objeto **AIMessage** con la misma estructura que las respuestas de otros proveedores:

````python
from langchain_google_genai import ChatGoogleGenerativeAI

modelo = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
respuesta = modelo.invoke("¿Qué es una API?")

# Contenido de la respuesta
print(f"Respuesta: {respuesta.content}")

# Metadatos de uso
print(f"Tokens de entrada: {respuesta.usage_metadata.get('input_tokens')}")
print(f"Tokens de salida: {respuesta.usage_metadata.get('output_tokens')}")
print(f"Total tokens: {respuesta.usage_metadata.get('total_tokens')}")
````

> Los tokens consumidos determinan el coste de cada petición. Monitorizar el uso de tokens te ayuda a controlar los gastos.

### Usar diferentes modelos

Puedes cambiar entre modelos simplemente modificando el parámetro `model`. Cada modelo tiene características diferentes en términos de velocidad, capacidad y coste:

````python
from langchain_google_genai import ChatGoogleGenerativeAI

pregunta = "Explica la teoría de la relatividad en términos simples"

# Modelo rápido para pruebas
modelo_flash = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
print("=== Gemini Flash ===")
print(modelo_flash.invoke(pregunta).content)

# Modelo avanzado para tareas complejas
modelo_pro = ChatGoogleGenerativeAI(model="gemini-2.5-pro")
print("\n=== Gemini Pro ===")
print(modelo_pro.invoke(pregunta).content)
````

### Conversaciones con mensajes estructurados

Los modelos de Google soportan todos los formatos de mensajes estándar de LangChain:

````python
from langchain_google_genai import ChatGoogleGenerativeAI

modelo = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# Usando tuplas con roles
mensajes = [
    ("system", "Eres un experto en historia del arte. Responde de forma concisa."),
    ("human", "¿Quién pintó la Mona Lisa?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

También puedes usar el formato diccionario:

````python
from langchain_google_genai import ChatGoogleGenerativeAI

modelo = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

mensajes = [
    {"role": "system", "content": "Eres un traductor al inglés."},
    {"role": "user", "content": "Buenos días, ¿cómo estás?"}
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

## Comparar Google con otros proveedores

Una ventaja de usar LangChain es que la **interfaz es idéntica** para diferentes proveedores. Esto permite cambiar entre modelos locales y en la nube con cambios mínimos en el código.

````python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_anthropic import ChatAnthropic
from langchain_openai import ChatOpenAI

pregunta = "¿Cuántos planetas tiene el sistema solar?"

# Modelo de Google
modelo_google = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
print("=== Google Gemini Flash ===")
respuesta_google = modelo_google.invoke(pregunta)
print(respuesta_google.content)
print(f"Tokens: {respuesta_google.usage_metadata.get('total_tokens')}")

# Modelo de Anthropic
modelo_anthropic = ChatAnthropic(model="claude-haiku-4-5-20251001")
print("\n=== Anthropic Claude Haiku ===")
respuesta_anthropic = modelo_anthropic.invoke(pregunta)
print(respuesta_anthropic.content)
print(f"Tokens: {respuesta_anthropic.usage_metadata.get('total_tokens')}")

# Modelo de OpenAI
modelo_openai = ChatOpenAI(model="gpt-5-mini")
print("\n=== OpenAI GPT-5-mini ===")
respuesta_openai = modelo_openai.invoke(pregunta)
print(respuesta_openai.content)
print(f"Tokens: {respuesta_openai.usage_metadata.get('total_tokens')}")
````

## Características distintivas de Gemini

### Capacidad multimodal

Gemini destaca por su capacidad de procesar múltiples tipos de contenido en una misma conversación: texto, imágenes, audio, video y PDFs. Esto lo convierte en una opción versátil para aplicaciones que requieren análisis de diferentes tipos de medios.

````python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.messages import HumanMessage

modelo = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# Analizar una imagen desde URL
mensaje = HumanMessage(
    content=[
        {"type": "text", "text": "Describe esta imagen en una frase."},
        {
            "type": "image",
            "url": "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/800px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg",
        },
    ]
)

respuesta = modelo.invoke([mensaje])
print(respuesta.content)
````

Los tipos de entrada soportados incluyen:

| Tipo | Formatos soportados |
|------|---------------------|
| Imágenes | URL, base64, Google Cloud Storage |
| Audio | URL, base64, archivos subidos |
| Video | base64, archivos subidos, YouTube (preview) |
| PDF | URL, base64, archivos subidos |

### Consideraciones de temperatura para Gemini 3.0+

Los modelos Gemini 3.0 y superiores tienen una particularidad importante: **la temperatura por defecto es 1.0** en lugar de 0.7. Usar valores más bajos con estos modelos puede causar problemas de rendimiento.

````python
from langchain_google_genai import ChatGoogleGenerativeAI

# Para Gemini 3.0+, temperature=1.0 es el valor recomendado
modelo = ChatGoogleGenerativeAI(
    model="gemini-3-pro-preview",
    temperature=1.0  # Valor recomendado para Gemini 3.0+
)

respuesta = modelo.invoke("Escribe un haiku sobre la programación")
print(respuesta.content)
````

> Si no especificas temperature, LangChain lo configurará automáticamente a 1.0 para modelos Gemini 3.0+.

### Thinking support

Algunos modelos Gemini soportan **thinking**, una funcionalidad que permite ajustar el presupuesto de tokens dedicados al razonamiento interno:

````python
from langchain_google_genai import ChatGoogleGenerativeAI

# Modelo con thinking habilitado
modelo = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    thinking_budget=1024,  # Tokens para razonamiento interno
    include_thoughts=True  # Incluir resumen del razonamiento en la respuesta
)

respuesta = modelo.invoke("Si tengo 3 manzanas y me dan el doble, ¿cuántas tengo?")
print(respuesta.content)
````

## Acceso alternativo: Vertex AI

Además del acceso directo a través de la API de Google AI (Gemini Developer API), los modelos Gemini también están disponibles a través de **Vertex AI**, la plataforma de machine learning de Google Cloud.

La misma clase `ChatGoogleGenerativeAI` puede conectarse a Vertex AI configurando las credenciales apropiadas:

````python
from langchain_google_genai import ChatGoogleGenerativeAI

# Usando Vertex AI en lugar de la API de Google AI
modelo = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    vertexai=True,
    project="tu-proyecto-gcp",
    location="us-central1"
)
````
